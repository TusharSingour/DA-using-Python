```python
a = 10
b = 5

print("Addition =", a + b)
print("Subtraction =", a - b)
print("Multiplication =", a * b)
print("Division =", a / b)

x = True
y = False

print("Logical AND =", x and y)
print("Logical OR =", x or y)
print("Logical NOT =", not x)
```

```python
# if, elif, else

num = int(input("Enter a number: "))

if num > 0:
    print("Positive Number")
elif num < 0:
    print("Negative Number")
else:
    print("Zero")

# for loop

print("For Loop:")
for i in range(1, 6):
    print(i)

# while loop

print("While Loop:")
count = 1
while count <= 5:
    print(count)
    count += 1
```

```python
# User Defined Function

def add(a, b):
    return a + b

result = add(10, 20)

print("Sum =", result)
```

```python
import math

print("Square Root =", math.sqrt(25))
```

```python
# List
my_list = [10, 20, 30, 40]
print("List:", my_list)

# Tuple
my_tuple = (1, 2, 3, 4)
print("Tuple:", my_tuple)

# Dictionary
my_dict = {"Name": "Tushar", "Age": 20}
print("Dictionary:", my_dict)

# Set
my_set = {10, 20, 30, 40, 20}
print("Set:", my_set)
```

```python
# String Operations

str1 = "Hello"
str2 = "Python"

# Concatenation
print("Concatenation:", str1 + " " + str2)

# Length
print("Length:", len(str1))

# Uppercase
print("Uppercase:", str1.upper())

# Lowercase
print("Lowercase:", str2.lower())

# Slicing
print("Slicing:", str2[0:3])
```

```python
try:
    num1 = int(input("Enter First Number: "))
    num2 = int(input("Enter Second Number: "))

    result = num1 / num2

    print("Result =", result)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid Input")
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1], [2], [3], [4], [5]])
Y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X, Y)

prediction = model.predict([[6]])

print("Prediction:", prediction)
```
